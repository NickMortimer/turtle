```python
import pandas as pd
import os
import geopandas as gp
```

```python
inputdir = '/home/mor582/seagrass'
country = 'TH'
cameratype = 'RGB'
dronetype='DJIP4'
surveycode='TKB'
surveynumber=5
timedif = '+07'
inputpath =os.path.join(inputdir,country,f'{surveycode}_{surveynumber:03}')
```

```python
data = pd.read_csv(f'{inputpath}/{dronetype}_{country}_{surveycode}_{surveynumber:03}_IMG_GP.CSV',
                   index_col='Time',parse_dates=['Time'])

area = gp.read_file(f'{inputpath}/{country}_{surveycode}_surveyarea.shp')
```

```python
from ipyleaflet import Map, basemaps, basemap_to_tiles, SplitMapControl



center = (data.iloc[0].Latitude, data.iloc[0].Longitude)
zoom = 15

map1 = Map(center=center, zoom=zoom, interpolation='nearest', basemap= basemaps.CartoDB.DarkMatter)

right_layer = basemap_to_tiles(basemaps.NASAGIBS.ModisTerraTrueColorCR, data.index[0].strftime('%Y-%m-%d'))

control = SplitMapControl(right_layer=right_layer)
map1.add_control(control)


map1
```

```python
from ipyleaflet import Map, TileLayer
planet_tiles_url=('https://tiles.planet.com/basemaps/v1/planet-tiles/global_monthly_'+ data.index[0].strftime('%Y_%m')+'_mosaic/gmap/{z}/{x}/{y}.png?api_key=38e806a52c8147b485003cf5e1aca6cd')
tile_layer=TileLayer(url=planet_tiles_url)
center = (data.iloc[0].Latitude, data.iloc[0].Longitude)
zoom = 15
m = Map(center = center,
    default_tiles=tile_layer,
    zoom=zoom,
    scroll_wheel_zoom=True)

m.add_layer(tile_layer)
#marker_cluster = MarkerCluster(markers = markers)
def add_make_dot(m,lat,lon,inarea):
    circle = Circle()
    circle.location = (lat,lon)
    circle.radius = 5
    if inarea:
        circle.color = "green"
        circle.fill_color = "green"
    else:
        circle.color = "red"
        circle.fill_color = "red"        
    circle.opacity = 1
    circle.fill_opacity = 1
    circle.stroke = False
    m.add_layer(circle)

from ipyleaflet import Map, Polygon
poly =list(map(lambda x: (x[1],x[0]),list(area.to_crs(epsg=4326).iloc[0].geometry.exterior.coords)))
polygon = Polygon(
    locations=poly,
    color="green",
    fill_color="green"
)
    
    
markers = [add_make_dot(m,item.Latitude,item.Longitude,item.InArea) for index,item in data.iterrows()]
m.add_layer(polygon)
m
```
